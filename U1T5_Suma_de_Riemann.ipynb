```python
import numpy as np   # cálculo numérico
import matplotlib.pyplot as plt  # graficación
```

# **Definiendo función para graficar**

**Función para graficar**

```python
def graficar(f, a, b, n):

  # generando un arreglo para los valores de x
  x = np.linspace(a,b,10*n)
  # generando un arreglo para los valores de y
  y = f(x)

  # generando un arreglo con la partición del intervalo (a,b)
  xi = np.linspace(a,b,n+1) # genera arreglo (a,b) con un tamaño de paso n+1
  # generando un arreglo con las alturas para cada valor xi
  yi = f(xi)
  # incremento de x
  base = (b-a)/n

  # PUNTOS EXTREMOS IZQUIERDOS
  x_Ln = xi[:-1] # todos los puntos menos el último
  y_Ln = yi[:-1]

  S = sum(base*y_Ln)  # suma rectángulos


  plt.figure(figsize=(15,5)) # definiendo la figura
  plt.subplot(1,3,1) # posición de la primera gráfica
  plt.plot(x,y,'b')  # grafica de la función

  plt.plot(x_Ln,y_Ln,'b.',markersize=10) # punto grande
  plt.bar(x_Ln,y_Ln, width = base, alpha=0.2, align='edge', edgecolor='b')
  plt.title('Suma de Riemann Ln = {:.2f}, n = {}'.format(S,n))

  # PUNTOS MEDIOS
  x_Mn = (xi[:-1] + xi[1:])/2 # puntos medios
  y_Mn = f(x_Mn)   # altura de puntos medios
  S = sum(base*y_Mn)  # suma rectángulos

  plt.subplot(1,3,2)
  plt.plot(x,y,'b')

  plt.plot(x_Mn,y_Mn, 'b.', markersize=10)
  plt.bar(x_Mn, y_Mn, width = base, alpha=0.2, edgecolor='b')
  plt.title('Suma de Riemann Mn = {:.2f}, n = {}'.format(S, n))

  # PUNTOS EXTREMOS DERECHOS
  x_Rn = xi[1:] # Right endpoints
  y_Rn = yi[1:]
  S = sum(base*y_Rn)  # suma rectángulos

  plt.subplot(1,3,3)
  plt.plot(x,y,'b')

  plt.plot(x_Rn,y_Rn,'b.',markersize=10)
  plt.bar(x_Rn,y_Rn,width = -base,alpha=0.2, align='edge', edgecolor='b')
  plt.title('Suma de Riemann Rn = {:.2f}, n = {}'.format(S,n))
  plt.show()
```

**Definiendo la función y el intervalo**

```python
a = 0
b = 1
n = 5
#f = lambda x : 1/(1+x**2) # definiendo la función
f = lambda x :x**2
# llamando la rutina para graficar
graficar(f,a,b,n)
```
